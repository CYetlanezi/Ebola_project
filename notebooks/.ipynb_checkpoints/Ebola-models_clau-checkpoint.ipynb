##  Ebolavirus

En 2014, África Occidental se enfrentó al brote de la enfermedad por el virus del Ébola (EVE) más grandede la historia. En agosto de 2018, se reportó un nuevo brote de ébola en la República Democrática del Congo con más de tres mil casos.

La enfermedad por el virus del Ébola (EVE) es una fiebre hemorrágica viral causada por Ebolavirus. Se identificó por primera vez en 1976 en Zaire1y su nombre se deriva del río Ébola ubicado cerca de la fuente del primer brote [6]. No fue hasta diciembre de 2019 que dos tratamientos (REGN-EB3 y mAb114) fueron asociadoscon mejores resultados. Los síntomas suelen comenzar entre dos días a tres semanas después de contraerlo. Los síntomas comienzan con fiebre, dolor de garganta, dolor muscular y dolores de cabeza. Siguen vómitos, diarrea y erupciones cutáneas y terminan con una disminución en la función del hígado y los riñones, provocando sangradointerno y externo. La enfermedad mata, en promedio, al 50 % de los infectados.

Una vez que el virus no se transmite por el aire o el agua, detener la transmisión del Ébola debería ser factible si se detecta a tiempo. Sin embargo, se ha demostrado que el Ébola se propaga por el aire en condiciones delaboratorio cuidadosamente controladas. Los entierros inseguros que involucran contacto directo con cuerpos infectados  con ébola  también  presentan  un  riesgo  de  infección  importante. Los problemas económicos, sociales y políticos en el Congo dificultan que adquiera vacunas y se posicione en un estado salubre en el cuál la enfermedad se erradique.

## Modelo 1. SIR

* **Susceptible.** 
* **Infectious.**
* **Recovered.** 

$$\text{Susceptible}
\xrightarrow{\frac{\beta S I}{N}} 
\text{Infectious} 
\xrightarrow{\gamma I} 
\text{Recovered} $$

\begin{align*}
\frac{dS}{dt} & = -\frac{\beta S I}{N} \\
\frac{dI}{dt} & = \frac{\beta S I}{N} - \gamma I \\
\frac{dR}{dt} & = \gamma I
\end{align*} 

### Parámetros

* $\beta$ = tasa de infección
* $\gamma$ = tasa de recuperación

## Unidades

* $\beta$ $\xrightarrow{}$ $[\frac{1}{tiempo}]$
* $\gamma$ $\xrightarrow{}$ $[\frac{1}{tiempo}]$

## Propiedades

* Los parametros $\beta$ y $\gamma$ tienen unidades 1 sobre tiempo.  
* $\beta$ es una tasa (cte) asociada con la transmición de la enfermedad. 
* La constante (temporal) $\tau_{infect} = \frac{1}{\beta}$ corresponde a el crecimiento exponencial de nuevas infecciones en una población en la que inicialmente los suceptibles son el total de la población ($S=N$). 
* $\gamma$ es la tasa de recuperación de los infectados. 
* La cte temporal asociada $\tau_{recovery}=\frac{1}{\gamma}$ es el tiempo promedio de recuperación de la infección.
* La población de infectados sólo puede crecer si y sólo si:
$\frac{dI}{dt} > 0 $  $\xrightarrow{}$  $\frac{\beta S I}{N} - \gamma I > 0$ $\xrightarrow{}$ $\frac{\beta S I}{N} > \gamma I $ $\xrightarrow{}$ $\frac{\beta S}{N} > \gamma $
* El radio $R_0 = \frac{\beta}{\gamma}$ es el número básico de reproducción que describe la transmisibilidad de la transmisión del virus.
* $R_0$ es el número promedio de personas infectadas por un caso al tiempo cero en una población completamente suceptible. Se puede ver como el tiempo promedio entre contactos es $\tau_{infect} = \frac{1}{\beta}$ y el tiempo promedio de eliminación es $\tau_{recovery}=\frac{1}{\gamma}$, de ahí se sigue que en promedio el número de contactos de una persona infectada con otras antes de que la enfermedad desaparezca es: $\frac{\tau_{infect}}{\tau_{recovery}}$
* La población infectada crece sólo si:  $\frac{\beta S}{N} > \gamma $ $\xrightarrow{}$ $\frac{\beta S}{\gamma N} > 1 $. Como inicialmente $S=N$, entonces  $R_0 > 1$ es suficiente para que la población infectada crezca.
* La población de infectados decrece si y sólo si:
$\frac{dI}{dt} < 0 $  $\xrightarrow{}$  $\frac{\beta S I}{N} - \gamma I < 0$ $\xrightarrow{}$ $\frac{\beta S I}{N} < \gamma I $ $\xrightarrow{}$ $\frac{\beta S}{N} < \gamma $ $\xrightarrow{}$ $\frac{\beta S}{\gamma N} < \gamma $ $\xrightarrow{}$ $\frac{R_0 S}{N} < 1 $ $\xrightarrow{}$ $\frac{S}{N} < \frac{1}{R_0} $.
* Como inicialmente $S=N$, entonces  $R_0 < 1$ es suficiente para que la población infectada decrezca.
* **La población ha alcanzado la inmunidad de rebaño cuando la fracción de suceptiblels es menor que $\frac{1}{R_0} $**
* Una vez que $\frac{dI}{dt} + \frac{dS}{dt} + \frac{dR}{dt} = 0 $ se tiene que integrando respecto del tiempo: $I(t) + S(t)+ R(t) = constante = N $ 

## Simulación

El siguiente código simula un model SIR y se extrajo de [[2]](https://jckantor.github.io/CBE30338/03.09-COVID-19.html). Los parámetros se obtuvieron de diversos artículos, revisar el detalle en el paper.

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# parameter values
R0 = 0.8 #
t_infective = 1/0.05#1/10-7 a 1/.10

# initial number of infected and recovered individuals
i_initial = 80/5620000
r_initial = 0.00
s_initial = 1 - i_initial - r_initial

gamma = 1/t_infective
beta = R0*gamma
```

```python
# SIR model differential equations.
def deriv(x, t, beta, gamma):
    s, i, r = x
    dsdt = -beta * s * i
    didt = beta * s * i - gamma * i
    drdt =  gamma * i
    return [dsdt, didt, drdt]

t = np.linspace(0, 60, 2000)
x_initial = s_initial, i_initial, r_initial
soln = odeint(deriv, x_initial, t, args=(beta, gamma))
s, i, r = soln.T
e = None

def plotdata(t, s, i, e=None):
    # plot the data
    fig = plt.figure(figsize=(12,6))
    ax = [fig.add_subplot(221, axisbelow=True), 
          fig.add_subplot(223),
          fig.add_subplot(122)]

    ax[0].plot(t, s, lw=3, label='Fraction Susceptible')
    ax[0].plot(t, i, lw=3, label='Fraction Infective')
    ax[0].plot(t, r, lw=3, label='Recovered')
    ax[0].set_title('Susceptible and Recovered Populations')
    ax[0].set_xlabel('Time /days')
    ax[0].set_ylabel('Fraction')

    ax[1].plot(t, i, lw=3, label='Infective')
    ax[1].set_title('Infectious Population')
    if e is not None: ax[1].plot(t, e, lw=3, label='Exposed')
    ax[1].set_ylim(0, 0.3)
    ax[1].set_xlabel('Time /days')
    ax[1].set_ylabel('Fraction')

    ax[2].plot(s, i, lw=3, label='s, i trajectory')
    ax[2].plot([1/R0, 1/R0], [0, 1], '--', lw=3, label='di/dt = 0')
    ax[2].plot(s[0], i[0], '.', ms=20, label='Initial Condition')
    ax[2].plot(s[-1], i[-1], '.', ms=20, label='Final Condition')
    ax[2].set_title('State Trajectory')
    ax[2].set_aspect('equal')
    ax[2].set_ylim(0, 1.05)
    ax[2].set_xlim(0, 1.05)
    ax[2].set_xlabel('Susceptible')
    ax[2].set_ylabel('Infectious')

    for a in ax: 
        a.grid(True)
        a.legend()

    plt.tight_layout()
    
```

## Modelo 2. Cuarentena, vacunación e inestabilidad socio política en la propagación del Ébola (Mayo 2020)

En esta sección analizaremos el modelo propuesto que consiste de seis compartimentos: Susceptibles (S),personas en cuarentena (Q), personas infectadas (I), personas hospitalizadas (H), personas recuperadas (R) y cadáveres infecciosos (D). Además, este modelo incorpora la variable $\omega$ parámetro que mide el impacto de la inestabilidad política de los países en la lucha contra la enfermedad.

* **Susceptibles (S).** 
* **Cuarentena (Q).**
* **Infectadas (I).**
* **Hospitalizadas (H)** 
* **Recuperadas (R)** 
* **Cadáveres infecciosos (D).** 

![A test image](model_2020.png)

$\frac{dS}{dt} = \pi - (\lambda + v + \mu + q) S$ 

$\frac{dQ}{dt} = qS - (1-\omega) \delta_Q Q - \mu Q$ 

$\frac{dI}{dt} = \lambda S - ((1-\omega) \delta_I + \gamma_I + \sigma_I + \mu) I$ 

$\frac{dH}{dt} = (1-w) (\delta_Q Q + \delta_I I) - (\sigma_H + \gamma_H + \mu) H$ 

$\frac{dR}{dt} = v S + \gamma_I I + \gamma_H H - \mu R$ 

$\frac{dD}{dt} = \sigma_I I + \sigma_H H - \rho D$

con $\lambda = \frac{\beta (I + \alpha_H H + \alpha_M D)}{N}$.
